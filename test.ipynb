```python
import numpy as np

img_array = np.asarray(img)
img_array.shape
```

```python
img_array = img_array[60:-25, :, :]
Image.fromarray(img_array)
```

```python
img_array = cv2.cvtColor(img_array, cv2.COLOR_RGB2YUV)
Image.fromarray(img_array)
```
